# Support Vector Machines

A support vector machine or SVM is a supervised learning algorithm used for classification tasks. It works on the principle of creating a "decision boundary" that can seperate an n-dimensional space into classes. Once the space is split into classes, it is easy to predict where an external or unseen data point lies amongst those classes.

This decision boundary that divides the space is called a "Hyperplane" and the points closest to the hyperplane are called the support vectors. The distance between the hyperplane and the support vector is called a "margin".




The purpose of the SVM algorithm is to maximize the margin or the distance between the hyperplane and the support vectors in order to find the optimal hyperplane that can classify the data points correctly.

## Hyperplane

A hyperplane is an (n-1) dimensional subspace which acts as the decision boundary between classes in an SVM, the optimization of which will give us seperable classes. Breaking the previous complicated line down - hyperplane being (n-1) dimensions refers to it having one less dimension than the amount of feature dimensions (n) - to simplify further, if the features are 2, then we are dealing with a 2 dimensional space and the hyperplane will be 2 - 1 = 1 dimensional, which is a line. This is exactly what we see in the image above. Similarly if the feature dimensions are a 3D space, the hyperplane will be a 2D plane. 

A hyperplane is said to be an optimized hyperplane when the margin from the support vectors is optimized.

A hyperplane is called a seperating hyperplane when it can successfully divides groups of data points which are linearly seperable in an ideal scenario. By linearly seperable, we mean a straight line can divide the two classes.

### Non-Linearly Seperable Data.

In an ideal scenario, the datapoints of different classes are scattered in such a way that they are linearly seperable. However, the ideal scenario does not usually occur in data and the classes are not directly linearly seperable.

In such cases, we employ various methods to try and transform the data itself, either by dimensional reduction or addition - or we try to ignore a few datapoints that cause the data to be non-linear.

For this, two major approaches can be taken into consideration:

- Soft Margin Approach
- Kernel Trick


### Soft Margin Approach

It is a method in which a few misclassifications of datapoints are tolerated to circumvent non-ideal data. This tolerance level is given by a degree of tolerance parameter. The degree of tolerance defines how many datapoints the algorithm will allow to be misclassified and how much the SVM will be penalized when it misclassifies a datapoint.

The higher the value of C, the shorter the margin value becomes.



### Hard Margin Approach

It is a method opposite to the soft margin, where misclassification of datapoints is not tolerated. This is usually not used since it caters well to more idealistic data, and such data is not commonplace. It has a degree of tolerance which is zero or close to zero and high penalization for misclassification. A hard margin however, gives the best and distinct decision line if the data is linearly seperable.

### Kernel Tricks

SVMs are simple and powerful tools for classification because the optimized hyperplane gives a clear boundary between classes. As we discussed before, in the case of linearly seperable data, SVMs work really well because it's easy to draw a hyperplane through the seperation. However, most real world data is noisy or simply non-linear in nature, and this is where the SVM starts to encounter problems and struggles to handle such data.

A solution to this problem is by applying a non-linear transformation to the datapoints which explicitly maps the input data to the higher-dimensional feature space using non-linear mapping. The transformation can be any non-linear mapping function that suits our problem (eg, polynomial). After the transformation is applied, the datapoints can be linearly seperated. 

This however, has one major problem - complexity - which increases exponentially with the amount of data since every single datapoint is being mapped explicitly to a higher dimension. This highly increases the computational cost. To resolve both the problem of computational complexity by explicit computation and non-linearity of real-world datasets - computer scientists came up with the 'kernel function'. 

Now a kernel function (or simply a kernel) is what allows us to perform what is known as the 'kernel trick'.  Simply put, the kernel function is capable of capturing the dissimilarity between data points in a feature space. This allows it to represent each transformed data point without explicitly computing what each data point is mapped to in the higher-dimension space, hence reducing computational cost.

Mathematically speaking, a kernel trick involves using the kernel function to represent the dot product between transformed feature vectors in the higher-dimensional space, without the need to actually explicitly compute the feature vectors in the higher dimensional space. This is represented by : 

$$ K(X_1,X_2) = X_1 ^T X_2 $$

where X1 and X2 are the input vectors.

The equation above is also called the linear kernel. Through this equation we understand that a dot product of the feature vectors in the higher dimensional space is simply the dot product of the original input vectors. By substitution of the kernel function into the SVM method, the need for any explicit computation for the mapping of datapoints to a higher dimensional feature space can be eliminated. A linear kernel however, will always provide a straight line - but we can replace the linear kernel function with other popular and optimized kernel functions.
