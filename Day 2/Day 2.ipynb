# Exploratory Data Analysis - The basics

Exploratory data analysis or EDA can be termed as finding patterns insights in the spread of data and exploring how features interact through various techniques and methods. EDA helps to uncover the underlying structure and characteristics of the data, which can guide further analysis and decision-making.

EDA is an unconstrained process. There are no fixed methods or processes for doing EDA on any dataset and it varies according to the type of data, the task at hand, and various other constraints like time.

There are various libraries that are used for EDA like `matplotlib` , `pandas`, `seaborn` , ```numpy``` and more.

### Barplot

A barplot or bar graph is an effective way of representing frequencies of categorical data as well as visualizing the relationship between the categorical data. A bar graph can be used to display the categories on the x-axis and the frequency/values/inputs belonging to that category on the y-axis.

Bar graphs allow for easy comparison of data across different groups or subgroups. By grouping the data based on a categorical variable, such as time periods, regions, or demographics, we can compare the distribution or values of another variable within each group using bar graphs. 

````python
sns.histplot(data = data, x = 'carbody')
style.use('fivethirtyeight')
````

### Histogram

Histograms are frequency charts that allow us to quickly and efficiently visualize information. They are an effective method to represent both categorical and continuous variables. Continuous variables can be made into categories by isolating the range into a number of 'bins' or subgroups.

A histogram may indicate the distribution of various features in the data, as well as the skewness of the data - as in if the data is skewed to the left or right.

````python
data['aspiration'].value_counts()
````

````python
b, bins, patches = plt.hist(
    [data.loc[data['aspiration'] == 'std', 'price'],
     data.loc[data['aspiration'] == 'turbo', 'price']],
    stacked=True,
    label=['Standard', 'Turbo'],
    edgecolor='white',
    bins = 8)
plt.legend()
````

From the histogram above, we can see that the price data is skewed to the left - this means we have more data for lower priced vehicles than the higher ones.

Interestingly, only a few cars have turbo aspirated engines. 

````python
b, bins, patches = plt.hist(
    [data.loc[data['aspiration'] == 'std', 'carbody'],
     data.loc[data['aspiration'] == 'turbo', 'carbody']],
    stacked=True,
    label=['Standard', 'Turbo'],
    edgecolor='white')
plt.legend()
````

Here, we can see that there's only a small number of examples that are attributed to the convertible, hardtop and wagon car body categories.

### Correlation Matrices

A correlation matrix is a tabular representation of the correlation coefficients between multiple variables in a dataset. It is a square matrix where each element represents the correlation between a pair of variables. The correlation coefficient measures the strength and direction of the linear relationship between two variables. Commonly used correlation coefficients include the Pearson correlation coefficient.

Correlation matrices allow us to identify multicollinearity - or the variables which are strongly correlated to the target variable, in this case say - price of the vehicle.

### Pearson Correlation Coefficient

The pearson correlation coefficient is a statistical measure that quantifies the strength and direction of the linear relationship between two continuous variables. It is denoted by the symbol "r" and has values between -1 and 1.

$$
r = \frac{\sum{((x - μx) * (y - μy))) }}{\sum{(n * σx * σy)}}
$$
